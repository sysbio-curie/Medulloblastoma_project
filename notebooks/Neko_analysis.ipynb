```python
from yfiles_jupyter_graphs import GraphWidget
from typing import Dict
```

```python
w = GraphWidget()
```

```python
# filling w with nodes
objects = []
for idx, item in group1_net.nodes.iterrows():
    obj = {
        "id" : group1_net.nodes["Genesymbol"].loc[idx],
        "properties" : {"label": group1_net.nodes["Genesymbol"].loc[idx]},
        "color":"#ffffff",
        "styles":{"backgroundColor":"#ffffff"}
        }
    objects.append(obj)
w.nodes=objects
```

```python
edges_WNT = group1_net.edges.fillna('')
```

```python
objects= []
for index, row in edges_WNT.iterrows():
    obj={
    "id":edges_WNT["Effect"].loc[index],
    "start" : edges_WNT["source"].loc[index],
    "end" : edges_WNT["target"].loc[index],
    "properties":{"references":edges_WNT["References"].loc[index]}}
    objects.append(obj)
w.edges=objects
```

```python
def custom_edge_color_mapping(edge: Dict):
    """let the edge be purple if the starting node has an even index"""
    if edge['id']  == "inhibition":
        color="#ff0066"
    elif edge['id']  == "stimulation":
        color="#00cc00"
    elif edge['id']  == "bimodal":
        color="#ffa500"
    elif edge['id']  == "undefined":
        color="#d834eb"
    return color
w.set_edge_color_mapping(custom_edge_color_mapping)


def custom_label_styles_mapping(node: Dict):
    """let the label be the negated purple big index"""
    return {
        'backgroundColor': '#ffffff', 
        'color': '#ffffff', 
        'shape':'circle',
        "fontSize": 200
    }
w.set_node_styles_mapping(custom_label_styles_mapping)
```

```python
w.directed=True
w.graph_layout='hierarchic'
w.show()
```
